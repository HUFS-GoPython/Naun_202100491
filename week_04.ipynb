### Frequency Distribution
- 단어의 빈도수를 활용할 수 있는 많은 메소드 제공
- 빈도수가 중요한 이유
  -인간 언어 -> 컴퓨터가 인식
  - 전체 단어 수 
  - 특정 단어의 빈도수

```python
fdist = nltk.FreqDist(text1)
#fdist # dictionary = key:value
```

```python
fdist.most_common(10)
```

```python
fdist['whale'] , text1.count('whale')
```

```python
fdist.freq('whale') #전체 택스트에서의 비율
```

```python
fdist.plot(50, cumulative=True)
```

```python
fdist.plot(50, cumulative=False)
```

```python
#1번 나타나는 단어들
#fdist.hapaxes()
len(fdist.hapaxes())  #중복 안됨, 대소문자 구분됨
```

```python
len(set(fdist.hapaxes()))
```

#### list comprehension
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성하는 것
- list + for_loop 한 줄로 해결하는 방식
- [A for A' in B]
  - B에 있는 A'에서 A를 가진 리스트를 반환
  - B: 기존의 리스트
  - A': B의 자료 형태
  - A: 새로운 리스트의 요소

- 조건+ : [A for A' in B if <조건식>]
  - [A for A'in B if <조건식> and <조건식>, ...]

```python
drinks
```

```python
[d for d in drinks]
```

```python
[d[0] for d in drinks]
```

```python
#text1 알파벳 수가 10 이상인 단어
#text1.tokens
#[word for word in text1 if len(word) >=10]
#[d for d in text1 if len(d) >=10]
```

```python
# text4에서 'gnt'를 포함하는 단어 변환

[word for word in text4 if 'gnt' in word]
```

```python
[word for word in set(text4) if 'gnt' in word] # 중복 없앰 -> 연산
```

```python
set([word for word in text4 if 'gnt' in word]) # 연산 -> 중복 없앰
```

```python
[len(word) for (word,count) in fdist4.items() if len(word) >= 4 and (count >=5)]
```

### 출력 포매팅
- 1. f'str...{var}'
  - f, F 둘다 가능
- 2. % -> 옛날 방식
 - '...%s'%var
- 3. '...{}'.format(var)

```python
name = 'lee'
print(f'My name is {name}.')
```

```python
print(F'My name is {name}')
```

```python
print('My name is %s.'%name)
```

```python
print('My name is {}.'.format(name))
```

```python
age = 14
```

```python
print(f'My name is {name}, and I am {age} years old.')
```

```python
print('My name is {1}, and I am {0} years old.'.format(name,age))
```

###할당
- = 기호 사용
- 수학에서는 '양변이 같다'라는 의미로 사용되지만, 프로그래밍에서는 '할당하다'라는 의미가 추가된다.
- "오른쪽의 값을 왼쪽에 할당하다."
   - 1. 오른쪽에 있는 모든 것은 값을 가져야 한다. (초기화를 해야 한다.)
   - 2. 같은 변수에 다른 값을 넣을 수 있다.

### 변수명 정하기
- 나쁜 변수명: a, b, c, a1, b1 등
- 의미있는 이름으로 선언해야 한다.
- 소문자, 대문자, 숫자, 언더바를 사용한다.예) name, name2, my_name, Name, NAME


```python
name = 'juliette'
name1 = 'jessica'
my_name = 'Laura'
#myName 카멜표기법
print(name, name1, my_name, sep=',')
```

- 안되는 것
  - 숫자로 시작할 수 없다.
    예: 22name, 22_name
  - 예약어로 선언할 수 없다.
    예: def, False
